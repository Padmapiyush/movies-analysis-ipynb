```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a sample movie data
movies = {
    'genre': ['Drama', 'Comedy', 'Action', 'Thriller', 'Drama', 'Comedy', 'Drama', 'Action', 'Thriller', 'Comedy'],
    'release_year': [2010, 2015, 2020, 2017, 2012, 2018, 2008, 2019, 2016, 2014],
    'budget': [30, 20, 80, 50, 40, 15, 60, 90, 70, 25],
    'box_office': [100, 80, 150, 120, 90, 60, 130, 180, 140, 75]
}
```

```python
# Create DataFrame object
df = pd.DataFrame(movies)
```

```python
# Dominant genre
genre_counts = df['genre'].value_counts()
dominant_genre = genre_counts.idxmax()
percent_dominant_genre = genre_counts.max() / len(df) * 100
```

```python
# Average release year
avg_release_year = df['release_year'].mean()
std_release_year = df['release_year'].std()
```

```python
# Correlation between budget and box office
correlation_budget_box_office = df['budget'].corr(df['box_office'])
```

```python
# Print results
print(f"Dominant genre: {dominant_genre} ({percent_dominant_genre:.2f}%)")
print(f"Average release year: {avg_release_year:.2f} years (STD: {std_release_year:.2f} years)")
print(f"Correlation between budget and box office gross: {correlation_budget_box_office:.2f}")
```

```python
# Plot distribution of release years
plt.hist(df['release_year'])
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.title('Distribution of Release Years in Movie Data')
plt.show()
```

```python
# Plot budget vs box office
plt.scatter(df['budget'], df['box_office'])
plt.xlabel('Budget (Millions)')
plt.ylabel('Box Office Gross (Millions)')
plt.title('Budget vs Box Office Gross')
plt.show()
```
